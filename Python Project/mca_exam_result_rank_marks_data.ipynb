```python
top10_ranks = df.head(10)
top10_ranks
#The list of first 1 to 10 rank students scored the highest marks (decreasing manner arrangement) 
```

```python
plt.figure(figsize=(10,6))
plt.bar(top10_ranks['Name'],top10_ranks['Marks'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Candidate Name')
plt.ylabel('Marks')
plt.title('Top 10 students by marks')
plt.tight_layout()
plt.show()
```

```python
bottom10_ranks = df.tail(10)
bottom10_ranks
#The list of bottom 10 rank students scored the highest marks (decreasing manner arrangement) 
```

```python
plt.figure(figsize=(10,6))
plt.bar(bottom10_ranks['Name'],bottom10_ranks['Marks'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Candidate Name')
plt.ylabel('Marks')
plt.title('Bottom 10 students by marks')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Marks'],df['Rank'],color='teal')
plt.gca().invert_yaxis()
plt.title('Scatter plot of marks vs rank')
plt.xlabel('Marks')
plt.ylabel('Rank')
plt.grid()
plt.show()
```
